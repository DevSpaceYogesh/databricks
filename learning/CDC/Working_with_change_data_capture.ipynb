```sql
create catalog if not exists playground;
```

```sql
create schema if not exists playground.sports
```

```sql
ALTER TABLE playground.sports.soccer_games
SET TBLPROPERTIES (
  delta.enableChangeDataFeed = true
);
```

```sql
CREATE TABLE playground.sports.sales (
  id INT,
  product STRING,
  amount DOUBLE
)
USING DELTA
TBLPROPERTIES (
  delta.enableChangeDataFeed = true
);
```

```sql
SHOW TBLPROPERTIES playground.sports.sales ('delta.enableChangeDataFeed');
```

```sql
INSERT INTO playground.sports.sales VALUES (1,'Laptop',1000);
```

```sql
UPDATE playground.sports.sales
SET amount = 1200
WHERE id = 1;
```

```sql
DELETE FROM playground.sports.sales
WHERE id = 1;
```
