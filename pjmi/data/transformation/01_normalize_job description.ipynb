```python
from pyspark.sql.functions import regexp_extract

df_rawjobs=spark.read.table("pjmi.bronze.raw_jobs")

df_jobdetails=(df_rawjobs.withColumn("JobTitle",regexp_extract("JobText",r"(.*?)\n",1))
               .withColumn("Experience_min",regexp_extract("jobText",r"(\d+)\s*-\s*(\d+)\s*years",1))
               .withColumn("Experience_max",regexp_extract("jobText",r"(\d+)\s*-\s*(\d+)\s*years",2))
               .withColumn("Location",regexp_extract("jobText",r"\n([^\n]+)\nSend me jobs like this",1))
               .withColumn("Application_nos",regexp_extract("jobText",r"Applicants:\s*([^\n]+)\n",1))
               .withColumn("Openings_no",regexp_extract("jobText",r"Openings:\s*(\d+)\n",1))
               .withColumn("Role",regexp_extract("jobText",r"Role:\s*(.*?)\n",1))
               .withColumn("Company",regexp_extract("jobText",r"\n(.*?)\n(\d+)",1)))


df_jobdetails.display()

```
